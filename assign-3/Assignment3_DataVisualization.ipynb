**1. Setup**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv("train_cleaned.csv")
```

**2. Class Task – Basic Plots**

a) Bar Chart

```python
plt.figure(figsize=(6,4))
df['MSZoning'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Count of Properties by Zoning')
plt.xlabel('Zoning')
plt.ylabel('Count')
plt.show()
```

* RL is the most common zoning type.

b) Scatter Plot

```python
plt.figure(figsize=(6,4))
plt.scatter(df['GrLivArea'], df['SalePrice'], alpha=0.5)
plt.title('Living Area vs Sale Price')
plt.xlabel('Above Ground Living Area (sq ft)')
plt.ylabel('Sale Price')
plt.show()
```

* Larger living area generally increases price, but a few extreme outliers exist.

c) Histogram

```python
plt.figure(figsize=(6,4))
df['SalePrice'].hist(bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

* Prices are right-skewed; most houses are clustered below $200k.

**3. Assignment 3 – Create 5 Different Plots**

**a. Boxplot – SalePrice by Overall Quality**

```python
plt.figure(figsize=(7,5))
sns.boxplot(
    x='OverallQual', 
    y='SalePrice', 
    hue='OverallQual',
    legend=False,
    data=df, 
    palette='coolwarm'
)
plt.title('Sale Price vs Overall Quality')
plt.show()
```

* Higher quality ratings correspond to higher median sale prices.

️**b. Heatmap – Feature Correlation**

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False).to_frame(),
            annot=True, cmap='YlOrRd')
plt.title('Correlation of Features with Sale Price')
plt.show()
```

* OverallQual and GrLivArea are the strongest positive correlates with price.

**c. Countplot – Garage Type**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='GarageType', data=df, order=df['GarageType'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Number of Houses by Garage Type')
plt.show()
```

* Attached garages dominate the market.

**d. Pairplot – Key Numeric Features**

```python
sns.pairplot(df[['SalePrice','GrLivArea','TotalBsmtSF','OverallQual']], diag_kind='kde')
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```

* Living area and basement size both rise with sale price.

**e. KDE Plot – Distribution of Lot Area**

```python
plt.figure(figsize=(6,4))
sns.kdeplot(df['LotArea'], fill=True, color='purple')
plt.title('Density of Lot Area')
plt.xlabel('Lot Area')
plt.show()
```

* Most properties have lot areas below 15,000 sq ft, with a long tail of larger lots.
